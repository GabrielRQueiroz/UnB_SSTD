```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
```

# Questão 5.6-7


```python
a = 0.9
k = 1.7 * (1 - a * np.sqrt(2) + a**2)

b0 = 1
b1 = -np.sqrt(2)
b2 = 1
a0 = 1
a1 = -np.sqrt(2) * a
a2 = a**2


w = np.linspace(-np.pi, np.pi, 10000)
w_axis = np.exp(1j * w)

H = k * (b0 * w_axis**2 + b1 * w_axis + b2) / (a0 * w_axis**2 + a1 * w_axis + a2)
H_abs = abs(H)
H_phase_deg = np.angle(H, deg=True)


fig, ax = plt.subplots(subplot_kw={"projection": "polar"})
omega = np.pi / 4
ax.set_rmax(1)
ax.grid(True)
# ZEROS
ax.plot(omega, 1, "ob", fillstyle="none", markersize=8, mew=2)
ax.plot(-omega, 1, "ob", fillstyle="none", markersize=8, mew=2)
# POLES
ax.plot(omega, a, "xb", fillstyle="none", markersize=8, mew=2)
ax.plot(-omega, a, "xb", fillstyle="none", markersize=8, mew=2)
plt.show()


plt.plot(w, H_abs)
plt.title("Resposta em Frequência de H(e^jw)")
plt.xlabel("Frequência angular $\omega$")
plt.ylabel("Magnitude")
plt.show()

plt.plot(w, H_phase_deg)
plt.title("Resposta em Frequência de H(e^jw)")
plt.xlabel("Frequência angular $\omega$")
plt.ylabel("Fase (Graus)")
plt.show()
```

```python
b0 = 0
b1 = 0.0783
b2 = 0
a0 = 1
a1 = -1.562
a2 = 0.6412


w = np.linspace(-np.pi, np.pi, 10000)
w_axis = np.exp(1j * w)

H = (b0 * w_axis**2 + b1 * w_axis + b2) / (a0 * w_axis**2 + a1 * w_axis + a2)
H_abs = abs(H)
H_phase_deg = np.angle(H, deg=True)

poles = np.roots([a0, a1, a2])

fig, ax = plt.subplots(subplot_kw={"projection": "polar"})
omega1 = np.log(poles[0])
omega2 = np.log(poles[1])
ax.set_rmax(1)
ax.grid(True)
# ZEROS
ax.plot(0, 0, "ob", fillstyle="none", markersize=8, mew=2)
# POLES
ax.plot(omega1, 1, "xb", fillstyle="none", markersize=8, mew=2)
ax.plot(omega2, 1, "xb", fillstyle="none", markersize=8, mew=2)
ax.plot(-omega1, 1, "xb", fillstyle="none", markersize=8, mew=2)
ax.plot(-omega2, 1, "xb", fillstyle="none", markersize=8, mew=2)
plt.show()


plt.plot(w, H_abs)
plt.title("Resposta em Frequência de H(e^jw)")
plt.xlabel("Frequência angular $\omega$")
plt.ylabel("Magnitude")
plt.show()

plt.plot(w, H_phase_deg)
plt.title("Resposta em Frequência de H(e^jw)")
plt.xlabel("Frequência angular $\omega$")
plt.ylabel("Fase (Graus)")
plt.show()
```
