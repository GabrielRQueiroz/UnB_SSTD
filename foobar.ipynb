```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
```

```python
alpha = 0.8

b0 = 0
b1 = 1
b2 = 1
a0 = 0
a1 = 1
a2 = - 1 * alpha

w = np.linspace(-np.pi, np.pi, 10000)
w_axis = np.exp(1j * w)

H = (b0 * w_axis**2 + b1 * w_axis + b2) / (a0 * w_axis**2 + a1 * w_axis + a2)
H_abs = abs(H)
H_phase_deg = np.angle(H, deg=True)

print(max(H_abs))

plt.plot(w, H_abs)
plt.title("Resposta em Frequência de H(e^jw)")
plt.xlabel("Frequência angular $\omega$")
plt.ylabel("Magnitude")
plt.show()

plt.plot(w, H_phase_deg)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Fase (Graus)')
plt.show()


#Note que é comum calcular a resposta em magnitude em dB e a resposta em fase em radianos
H_mag_db = 10 * np.log10(abs(H))
H_phase = np.angle(H)

plt.plot(w, H_mag_db)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Magnitude (dB)')
plt.show()

plt.plot(w, H_phase)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Fase (Radianos)')
plt.show()
```
